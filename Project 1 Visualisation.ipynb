```python
# Import necessary libraries
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
%matplotlib notebook
import hvplot.pandas

# Specify the path to your CSV file
csv_file = Path("C:\\Users\\sooki\\OneDrive\\Desktop\\universal_top_spotify_songs (RAW).csv")

# Load the CSV file into a DataFrame
spotify_songs = pd.read_csv(csv_file)

# Preview the data
spotify_songs.head()
```

```python
country_selected = ["US", "CA", "BR", "AR", "GB", "DE", "KR", "JP", "NG", "ZA", "AU", "NZ"]

filtered_data = spotify_songs[spotify_songs["country"].isin(country_selected)]

filtered_data
```

```python
top_artists = filtered_data["artists"].value_counts().head(10)
top_artists.head()
```

```python
plt.figure(figsize=(6, 6))
top_artists.plot(kind="bar", color="skyblue")
plt.title("Top 10 Artists within the Selected Countries")
plt.xlabel("Artist Name")
plt.ylabel("Count of Appearance")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

plt.savefig(r"C:\Users\sooki\OneDrive\Desktop\Testing123\Graph1.png")
```

```python
taylor_swift_data = filtered_data[filtered_data["artists"].str.contains("Taylor Swift", case=False, na=False)]

taylor_swift_counts = taylor_swift_data.groupby("country").size().reset_index(name="count")
taylor_swift_counts
```

```python
plt.figure(figsize=(7, 4))
plt.bar(taylor_swift_counts["country"], taylor_swift_counts["count"], color="skyblue")
plt.title("Count of Taylor Swift Songs across Selected Countries")
plt.xlabel("Selected Country")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right") 
plt.tight_layout()

plt.show()
plt.savefig(r"C:\Users\sooki\OneDrive\Desktop\Testing123\Graph2.png")
```

```python
reselected_countries = ["AR", "BR", "DE", "KR", "ZA"]

filtered_data_reselected = filtered_data[filtered_data["country"].isin(reselected_countries)]

# Find the top artists in the reselected countries
top_artists_in_reselected_countries = filtered_data_reselected["artists"].value_counts()
top_artists_in_reselected_countries.head(10)
```

```python
top_artists_in_reselected_countries = filtered_data_reselected["artists"].value_counts().head(10)

plt.figure(figsize=(7, 6))
top_artists_in_reselected_countries.plot(kind="bar", color="skyblue")
plt.title("Top 10 Artists in Selected Countries (where Taylor Swift is Less Popular)")
plt.xlabel("Artist Name")
plt.ylabel("Count of Appearance")
plt.xticks(rotation=45, ha="right") 
plt.tight_layout()

plt.show()

plt.savefig(r"C:\Users\sooki\OneDrive\Desktop\Testing123\Graph3.png")
```
